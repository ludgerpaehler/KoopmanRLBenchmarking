Define the universal index for all plots

```python
xs = skvi_fluid_flow_df.index
```

Calculate all means, and standard deviations for the plotting

Linear System:

```python
# Values for the plotting of the linear system

# LQR
ymean_linear_system_lqr = lqr_linear_system_df.mean(axis=1)
ystd_linear_system_lqr = lqr_linear_system_df.std(axis=1)

# SKVI
ymean_linear_system_skvi = skvi_linear_system_df.mean(axis=1)
ystd_linear_system_skvi = skvi_linear_system_df.std(axis=1)

# SAC (Q)
ymean_linear_system_sac_q = sac_q_linear_system_df.mean(axis=1)
ystd_linear_system_sac_q = sac_q_linear_system_df.std(axis=1)

# SAC (V)
ymean_linear_system_sac_v = sac_v_linear_system_df.mean(axis=1)
ystd_linear_system_sac_v = sac_v_linear_system_df.std(axis=1)

# SAKC
ymean_linear_system_sakc = sakc_linear_system_df.mean(axis=1)
ystd_linear_system_sakc = sakc_linear_system_df.std(axis=1)
```

Fluid Flow:

```python
# Values for the plotting of the fluid flow

# LQR
ymean_fluid_flow_lqr = lqr_fluid_flow_df.mean(axis=1)
ystd_fluid_flow_lqr = lqr_fluid_flow_df.std(axis=1)

# SKVI
ymean_fluid_flow_skvi = skvi_fluid_flow_df.mean(axis=1)
ystd_fluid_flow_skvi = skvi_fluid_flow_df.std(axis=1)

# SAC (Q)
ymean_fluid_flow_sac_q = sac_q_fluid_flow_df.mean(axis=1)
ystd_fluid_flow_sac_q = sac_q_fluid_flow_df.std(axis=1)

# SAC (V)
ymean_fluid_flow_sac_v = sac_v_fluid_flow_df.mean(axis=1)
ystd_fluid_flow_sac_v = sac_v_fluid_flow_df.std(axis=1)

# SAKC
ymean_fluid_flow_sakc = sakc_fluid_flow_df.mean(axis=1)
ystd_fluid_flow_sakc = sakc_fluid_flow_df.std(axis=1)
```

Lorenz System:

```python
# Values for the plotting of the Lorenz system

# LQR
ymean_lorenz_lqr = lqr_lorenz_df.mean(axis=1)
ystd_lorenz_lqr = lqr_lorenz_df.std(axis=1)

# SKVI
ymean_lorenz_skvi = skvi_lorenz_df.mean(axis=1)
ystd_lorenz_skvi = skvi_lorenz_df.std(axis=1)

# SAC (Q)
ymean_lorenz_sac_q = sac_q_lorenz_df.mean(axis=1)
ystd_lorenz_sac_q = sac_q_lorenz_df.std(axis=1)

# SAC (V)
ymean_lorenz_sac_v = sac_v_lorenz_df.mean(axis=1)
ystd_lorenz_sac_v = sac_v_lorenz_df.std(axis=1)

# SAKC
ymean_lorenz_sakc = sakc_lorenz_df.mean(axis=1)
ystd_lorenz_sakc = sakc_lorenz_df.std(axis=1)
```

Double Well:

```python
# Values for the plotting of the Double Well

# LQR
ymean_double_well_lqr = lqr_double_well_df.mean(axis=1)
ystd_double_well_lqr = lqr_double_well_df.std(axis=1)

# SKVI
ymean_double_well_skvi = skvi_double_well_df.mean(axis=1)
ystd_double_well_skvi = skvi_double_well_df.std(axis=1)

# SAC (Q)
ymean_double_well_sac_q = sac_q_double_well_df.mean(axis=1)
ystd_double_well_sac_q = sac_q_double_well_df.std(axis=1)

# SAC (V)
ymean_double_well_sac_v = sac_v_double_well_df.mean(axis=1)
ystd_double_well_sac_v = sac_v_double_well_df.std(axis=1)

# SAKC
ymean_double_well_sakc = sakc_double_well_df.mean(axis=1)
ystd_double_well_sakc = sakc_double_well_df.std(axis=1)
```
